```python
import sklearn
from sklearn.model_selection import train_test_split
import os
```

```python
dataset_dir = "D:/6TH SEM/DIP/Indian Food Images/Indian Food Images"
```

```python
classes = os.listdir(dataset_dir)
classes
```

```python
file_paths = []
labels = []
```

```python
for class_name in classes:
    # Construct the full path to the class directory
    class_dir = os.path.join(dataset_dir, class_name)
    # List all files in the class directory
    class_files = os.listdir(class_dir)
    # Iterate over each file in the class directory
    for file_name in class_files:
        # Construct the full path to the file
        file_path = os.path.join(class_dir, file_name)
        # Append the file path to the list
        file_paths.append(file_path)
        # Append the label (class name) to the labels list
        labels.append(class_name)
```

```python
print(file_paths)
print(labels)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
train_file_paths, test_file_paths, train_labels, test_labels = train_test_split(
    file_paths, labels, test_size=0.2, random_state=42)

# Split the training data into training and validation sets
train_file_paths, validation_file_paths, train_labels, validation_labels = train_test_split(
    train_file_paths, train_labels, test_size=0.2, random_state=42)

print("Number of training samples:", len(train_file_paths))
print("Number of validation samples:", len(validation_file_paths))
print("Number of test samples:", len(test_file_paths))
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
```

```python
# Load the pre-trained VGG16 model without the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
```

```python
# Freeze the weights of the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Add custom classification layers on top of the VGG16 base model
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
predictions = Dense(80, activation='softmax')(x)  # Assuming 80 classes for the Indian food dataset
```

```python
# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Print model summary
model.summary()
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define the source directory where your original dataset is located
source_dir = "D:/6TH SEM/DIP/Indian Food Images/Indian Food Images"

# Define the destination directory where you want to organize your dataset
destination_dir = "D:/6TH SEM/DIP/Indian Food Images/Dataset"

# Create directories for training, validation, and test data
train_dir = os.path.join(destination_dir, 'train')
validation_dir = os.path.join(destination_dir, 'validation')
test_dir = os.path.join(destination_dir, 'test')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Get a list of all subdirectories (classes) in the source directory
classes = os.listdir(source_dir)

# Split the dataset into training, validation, and test sets for each class
for class_name in classes:
    class_dir = os.path.join(source_dir, class_name)
    image_files = [os.path.join(class_dir, file) for file in os.listdir(class_dir) if file.endswith('.jpg')]
    
    train_files, test_files = train_test_split(image_files, test_size=0.2, random_state=42)
    train_files, validation_files = train_test_split(train_files, test_size=0.2, random_state=42)
    
    # Move images to respective directories
    for file in train_files:
        destination = os.path.join(train_dir, class_name)
        os.makedirs(destination, exist_ok=True)
        shutil.copy(file, destination)
    
    for file in validation_files:
        destination = os.path.join(validation_dir, class_name)
        os.makedirs(destination, exist_ok=True)
        shutil.copy(file, destination)
    
    for file in test_files:
        destination = os.path.join(test_dir, class_name)
        os.makedirs(destination, exist_ok=True)
        shutil.copy(file, destination)
```

```python
# Define ImageDataGenerator for data preprocessing and augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255)

# Flow training images in batches of 32 using train_datagen generator
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),  # Resize images to 150x150 pixels
    batch_size=32,
    class_mode='categorical'  # Assumes categorical labels
)

# Flow validation images in batches of 32 using validation_datagen generator
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Flow test images (optional)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)
```

```python
import os
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Define model checkpoint to save the best model during training
checkpoint_path = "best_model.keras"
checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
```

```python
# Train the model
history = model.fit(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=[checkpoint]
)
```

```python
# Load the best model from the checkpoint
best_model = load_model(checkpoint_path)
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = best_model.evaluate(test_generator)
print(f'Test loss: {test_loss}')
print(f'Test accuracy: {test_accuracy}')
```

```python
# Save the best model for deployment
best_model.save('food_recognition_model_new2.keras')
```

```python
def predict_food_item(image_path):
        # Load and preprocess the input image
        img = image.load_img(image_path, target_size=(150, 150))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array /= 255.  # Normalize pixel values

        # Perform prediction
        predictions = model.predict(img_array)
        
        # Get the predicted class index
        predicted_class_index = np.argmax(predictions[0])

        # Map the class index to the corresponding food item (assuming you have a class label mapping)
        predicted_food_item = class_labels[predicted_class_index]
        
         # Get the calorie value for the predicted food item
        predicted_calories = food_calories[predicted_food_item]

        print("Predicted Food Item:",predicted_food_item)
        print("Calories:", predicted_calories)
```
